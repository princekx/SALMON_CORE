```python
import os
import sys
import iris
import matplotlib.pyplot as plt
%matplotlib qt
```

```python
def read_harmonics(harfile):
    # read harmonic coefficients from netcdf file
    mm = iris.load_cube(harfile, 'mm')
    aa = iris.load_cube(harfile, 'aa')
    bb = iris.load_cube(harfile, 'bb')
    return mm, aa, bb
```

```python
varname = 'olr'
harfile = os.path.join('../data', f'{varname}.nr.79to01.m+3har.nc')
```

```python
harfile
```

```python
mm, aa, bb = read_harmonics(harfile)
```
